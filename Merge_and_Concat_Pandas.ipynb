```python
#Merge
```

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,4],"C":[15,16,17,18]})
pd.merge(var1,var2,on="A")
```

```python
pd.merge(var2,var1,on="A")
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var4 = pd.DataFrame({"A":[1,2,3,5],"C":[15,16,17,18]})
pd.merge(var3,var4,on="A")
```

```python
pd.merge(var3,var4,how="inner")
```

```python
pd.merge(var3,var4,how="left")
```

```python
pd.merge(var3,var4,how="right")
```

```python
pd.merge(var3,var4,how="outer")
```

```python
pd.merge(var3,var4,how="outer",indicator = True)
```

```python
var5 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var6 = pd.DataFrame({"A":[1,2,3,4],"C":[5,6,7,8]})
pd.merge(var5,var6, left_index = True , right_index = True)
```

```python
pd.merge(var5,var6, left_index = True , right_index = True, suffixes = ("name" , "python"))
```

```python
#Concat
```

```python
sr1 = pd.Series([1,2,3,4,5])
sr2 = pd.Series([11,21,31,41,51])
pd.concat([sr1,sr2])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
d2 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
pd.concat([d1,d2])
```

```python
pd.concat([d1,d2],axis=1)
```
